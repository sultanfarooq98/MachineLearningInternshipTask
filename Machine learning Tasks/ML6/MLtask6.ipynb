```python
import pandas as pd

# Load the dataset
data = pd.read_csv(r"D:\Work\Development\Machine learning Tasks\ML6\user_profiles_for_ads.csv")

# Check for null values
print(data.isnull().sum())

# Display column info
print(data.info())

# Display descriptive statistics
print(data.describe(include='all'))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], kde=True)
plt.title('Age Distribution')
plt.show()

# Distribution of Gender
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution')
plt.show()

# Distribution of Education Level
plt.figure(figsize=(10, 6))
sns.countplot(x='Education Level', data=data)
plt.title('Education Level Distribution')
plt.show()

# Distribution of Income Level
plt.figure(figsize=(10, 6))
sns.histplot(data['Income Level'], kde=True)
plt.title('Income Level Distribution')
plt.show()
```

```python
# Device usage patterns
plt.figure(figsize=(10, 6))
sns.countplot(x='Device Usage', data=data)
plt.title('Device Usage Patterns')
plt.show()

# User engagement with content and ads
plt.figure(figsize=(10, 6))
sns.histplot(data['Time Spent Online (hrs/weekday)'], kde=True)
plt.title('User Engagement with Content')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(data['Ad Interaction Time (sec)'], kde=True)
plt.title('User Engagement with Ads')
plt.show()

# Common interests among users
plt.figure(figsize=(10, 6))
sns.countplot(y='Top Interests', data=data, order = data['Top Interests'].value_counts().index)
plt.title('Most Common Interests Among Users')
plt.show()
```

```python
# Average time spent online on weekdays vs weekends
average_time_weekdays = data[data['Time Spent Online (hrs/weekday)'].isin(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])]['Ad Interaction Time (sec)'].mean()
average_time_weekends = data[data['Time Spent Online (hrs/weekend)'].isin(['Saturday', 'Sunday'])]['Ad Interaction Time (sec)'].mean()

print('Average time spent online on weekdays:', average_time_weekdays)
print('Average time spent online on weekends:', average_time_weekends)

# User engagement metrics
average_likes = data['Likes and Reactions'].mean()
# average_reactions = data['reactions'].mean()

print('Average Likes:', average_likes)
# print('Average Reactions:', average_reactions)

# Ad interaction metrics
ctr = (data['Click-Through Rates (CTR)'] / data['Click-Through Rates (CTR)']).mean()
conversion_rate = (data['Conversion Rates'] / data['Click-Through Rates (CTR)']).mean()
ad_interaction_time = data['Ad Interaction Time (sec)'].mean()

print('Click-Through Rate (CTR):', ctr)
print('Conversion Rate:', conversion_rate)
print('Average Ad Interaction Time:', ad_interaction_time)
```
